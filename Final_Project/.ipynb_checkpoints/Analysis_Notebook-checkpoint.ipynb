### Memory-Based Collaborative Filtering (Item-Item  & User-User)

There are two Memory-Based Collaborative Filtering approaches, which are User-User filtering and item-item filtering. The user-user filtering takes a particular user, find users that are similar to that user based on similarity of ratings, and recommend items that those similar users liked. And the item-item filtering will take an item, find users who liked that item, and find other items that those users or similar users also liked. It takes items and outputs other items as recommendations.

We will use 5-folds cross-validation to make predictions and evaluate the performance of both Item-Item Collaborative Filtering and User-User Collaborative Filtering. Before we do the cross validation, we first introduce how we make predictions and evaluate their performance, and define some functions we will use in this process.

#### Similarity

In each iteration of 5-folds cross validation, we will create a user-user matrix which you build from the entire dataset. Since we will split the data into testing and training we will need to create two 943 × 1682 matrices. The training matrix contains 80% of the ratings and the testing matrix contains 20% of the ratings.

After we built the user-user matrix you calculate the similarity and create a similarity matrix. The similarity values between items in Item-Item Collaborative Filtering are measured by observing all the users who have rated both items. For User-User Collaborative Filtering the similarity values between users are measured by observing all the items that are rated by both users. The distance metric we will use is cosine similarity, where the ratings are seen as vectors in n-dimensional space and the similarity is calculated based on the angle between these vectors. 

Cosine similiarity for users $a$ and $m$ can be calculated using the formula below, where we take dot product of the user vector uk and the user vector ua and divide it by multiplication of the Euclidean lengths of the vectors.

<center>$s_u^{cos}(u_k,u_a)
 = \frac{ u_k \cdot u_a }{ \left \| u_k \right \| \left \| u_a \right \| }
 = \frac{ \sum x_{k,m}x_{a,m} }{ \sqrt{\sum x_{k,m}^2\sum x_{a,m}^2} }$</center>


We calculate similarity between items $m$ and $b$ using formula:

<center>$s_u^{cos}(i_m,i_b)
 = \frac{ i_m \cdot i_b }{ \left \| i_m \right \| \left \| i_b \right \| }
 = \frac{ \sum x_{a,m} x_{a,b} }{ \sqrt{ \sum x_{a,m}^2 \sum x_{a,b}^2 } }$</center>
 
We can use the pairwise_distances function from sklearn to calculate the cosine similarity.

#### Prediction

In terms of making prediction, there are two different prediction formula for user-based CF and item-based CF given the similarity matrices we created.

For user-based CF, some users may tend always to give high or low ratings to all movies. The relative difference in the ratings that these users give is more important than the absolute values. For example, user $k$ gives 4 stars to his favourite movies and 3 stars to all other good movies. Suppose now that another user $t$ rates movies that he/she likes with 5 stars, and the movies he/she fell asleep over with 3 stars. These two users could have a very similar taste but treat the rating system differently. So we first look at the similarity between users $k$ and $a$ as weights that are multiplied by the ratings of a similar user a (corrected for the average rating of that user). Then we need to normalize it so that the ratings stay between 1 and 5 and, as a final step, sum the average ratings for the user that we are trying to predict. The the prediction formula would be like the following:

<center>$\hat{x}_{k,m} = \bar{x}_{k} + \frac{\sum\limits_{u_a} sim_u(u_k, u_a) (x_{a,m} - \bar{x}_{u_a})}{\sum\limits_{u_a}|sim_u(u_k, u_a)|}$</center>

For item-based CF, we don't need to correct for users average rating since query user itself is used to do predictions. So the prediction formula would be like the following:

<center>$\hat{x}_{k,m} = \frac{\sum\limits_{i_b} sim_i(i_m, i_b) (x_{k,b}) }{\sum\limits_{i_b}|sim_i(i_m, i_b)|}$</center>

We define the function 'predict' to make predictions.

```python
def predict(ratings, similarity, type='user'):
    if type == 'user':
        mean_user_rating = ratings.mean(axis=1)
        #You use np.newaxis so that mean_user_rating has same format as ratings
        ratings_diff = (ratings - mean_user_rating[:, np.newaxis])
        pred = mean_user_rating[:, np.newaxis] + similarity.dot(ratings_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
    elif type == 'item':
        pred = ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])
    return pred
```

#### Evaluation

In terms of evaluation, we find it is hard to define a True Postive or True negative since it is close related to the number of our recommendation we give. Instead, we will evaluate the accuracy of predicted rating, here we use Root Mean Squared Error (RMSE). If RMSE value is small and close to 0, it means we can approximate the original matrix pretty well. Therefore, our recommendation would be more accurate. The definition of RMSE is as the following:

<center>$\mathit{RMSE} =\sqrt{\frac{1}{N} \sum (x_i -\hat{x_i})^2}$</center>

We define the function 'rmse' to calculate RMSE.

```python
def rmse(prediction, ground_truth):
    prediction = prediction[ground_truth.nonzero()].flatten()
    ground_truth = ground_truth[ground_truth.nonzero()].flatten()
    return sqrt(mean_squared_error(prediction, ground_truth))
```

Then we use the previously defined function doing 5-folds cross validation to make predictions and evaluate the performance of both Item-Item Collaborative Filtering and User-User Collaborative Filtering.

```python
# Cross Validation
user_rmse = []
item_rmse = []
# Construct a kfolds object
kf = KFold(df.shape[0],n_folds=5,shuffle=True)
# Iterate through folds
for train_index, test_index in kf:
    train_data = df.iloc[train_index]
    test_data = df.iloc[test_index]
    
    #Create two user-item matrices, one for training and another for testing
    train_data_matrix = np.zeros((n_users, n_items))
    for line in train_data.itertuples():
        train_data_matrix[line[1]-1, line[2]-1] = line[3]

    test_data_matrix = np.zeros((n_users, n_items))
    for line in test_data.itertuples():
        test_data_matrix[line[1]-1, line[2]-1] = line[3]
    
    #Calculate 1-cosine similarity
    user_similarity = pairwise_distances(train_data_matrix, metric='cosine')
    item_similarity = pairwise_distances(train_data_matrix.T, metric='cosine')
    
    #Make predictions by using 'predict' function
    item_prediction = predict(train_data_matrix, item_similarity, type='item')
    user_prediction = predict(train_data_matrix, user_similarity, type='user')
    
    #Calculate RMSE
    user_rmse.append(rmse(user_prediction, test_data_matrix))
    item_rmse.append(rmse(item_prediction, test_data_matrix))
    
#Print out the RMSE    
print('User-based CF cross-valication RMSE: ' + str(np.mean(user_rmse)))
print('Item-based CF cross-valication RMSE: ' + str(np.mean(item_rmse)))
```

### Model-based Collaborative Filtering

Model-based Collaborative Filtering is based on matrix factorization, mainly as an unsupervised learning method for latent variable decomposition and dimensionality reduction. Matrix factorization can deal better with scalability and sparsity than Memory-based Collaborative Filtering. The goal of matrix factorization is to learn the latent preferences of users and the latent attributes of items from known ratings to then predict the unknown ratings through the dot product of the latent features of users and items. By doing matrix factorization to a very sparse matrix with a lot of dimensionsm, we can restructure the user-item matrix into low-rank structure, and represent the matrix by the multiplication of two low-rank matrices, where the rows contain the latent vector. We then fit this matrix to approximate the original matrix by multiplying the low-rank matrices together, which fills in the entries missing in the original matrix.

First, we take a look of the sparsity level of MovieLens dataset:

```python
sparsity=round(1.0-len(df)/float(n_users*n_items),3)
print('The sparsity level of MovieLens100K is ' +  str(sparsity*100) + '%')
```

The matrix factorization method we will use is Singular value decomposition (SVD). The general equation can be expressed as follows: $X=U \times S \times V^T$.


Given an $m×n$ matrix $X$, it can be factorized to $U$, $S$ and $V^T$:

-  $U$ is an $m×r$ orthogonal matrix, which represents the feature vectors corresponding to the users in the hidden feature space
-  $S$ is an $r×r$ diagonal matrix with non-negative real numbers on the diagonal
-  $V^T$ is an $r×n$ orthogonal matrix and the $V$ matrix represents the feature vectors corresponding to the items in the hidden feature space

Below is an example of approximating the original matrix by SVD with K=20.

```python
#Split the dataset into testing and training.
train_data, test_data = cv.train_test_split(df, test_size=0.2)

#Create two user-item matrices, one for training and another for testing
train_data_matrix = np.zeros((n_users, n_items))
for line in train_data.itertuples():
    train_data_matrix[line[1]-1, line[2]-1] = line[3]

test_data_matrix = np.zeros((n_users, n_items))
for line in test_data.itertuples():
    test_data_matrix[line[1]-1, line[2]-1] = line[3]

#get SVD components from normalized train matrix

user_ratings_mean = np.mean(train_data_matrix, axis = 1)
normal_train_data_matrix = train_data_matrix- user_ratings_mean.reshape(-1, 1)

u, s, vt = svds(normal_train_data_matrix, k = 20)
s_diag_matrix=np.diag(s)
X_pred = np.dot(np.dot(u, s_diag_matrix), vt) + user_ratings_mean.reshape(-1, 1)
print('User-based CF MSE: ' + str(rmse(X_pred, test_data_matrix)))
```

Then we do the cross-validation to find the optimal K value which minimizes RMSE.

```python
#Cross Validation

# Construct a kfolds object
kf = KFold(df.shape[0],n_folds=5,shuffle=True)
rmse_value = []

# Iterate through folds
for train_index, test_index in kf:
    train_data = df.iloc[train_index]
    test_data = df.iloc[test_index] 
    
    #Create two user-item matrices, one for training and another for testing
    train_data_matrix = np.zeros((n_users, n_items))
    for line in train_data.itertuples():
        train_data_matrix[line[1]-1, line[2]-1] = line[3]

    test_data_matrix = np.zeros((n_users, n_items))
    for line in test_data.itertuples():
        test_data_matrix[line[1]-1, line[2]-1] = line[3]

    k_range = range(0, 105, 5)[1:]
    for k_value in k_range:
        
        #Get SVD components from normalized train matrix
        user_ratings_mean = np.mean(train_data_matrix, axis = 1)
        normal_train_data_matrix = train_data_matrix- user_ratings_mean.reshape(-1, 1)
        u, s, vt = svds(normal_train_data_matrix, k = k_value)
        s_diag_matrix=np.diag(s)
        
        #Make prediction
        X_pred = np.dot(np.dot(u, s_diag_matrix), vt) + user_ratings_mean.reshape(-1, 1)
        rmse_value.append(rmse(X_pred, test_data_matrix))
    
rmse_value = np.array(rmse_value)
shape = (5, 20)
rmse_value = rmse_value.reshape(shape)
mean_rmse_value = np.mean(rmse_value,axis=0)


plt.plot(k_range, mean_rmse_value)
plt.show()
```

We define functino to calculate cosine similarity, sort by most similar and return the top N movies that also in our selected movies list.

```python
def top_cosine_similarity(low_dim_data, movie_name, top_n):
    movie_name = movie_name
    low_case_movie_name = movie_name.lower()
    if low_case_movie_name in lower_case_movie_dict:   
        index = Rating_df.columns.get_loc(lower_case_movie_dict[low_case_movie_name])
        movie_row = low_dim_data[:, index]
        magnitude = np.sqrt(np.einsum('ij, ij -> i', low_dim_data.T, low_dim_data.T))
        similarity = np.dot(movie_row, low_dim_data) / (magnitude[index] * magnitude)
        pre_sort_indexes = np.argsort(-similarity)
        #Delete the index of input movie itself
        sort_indexes = np.delete(pre_sort_indexes, 0)
        top_movie_id = Rating_df.columns[sort_indexes]
        frozen_selected_id = frozenset(selected_id)
        selected_top_movie_id = [x for x in top_movie_id if x in frozen_selected_id][:top_n]
        return selected_top_movie_id
    else:
        return("Sorry, no matching found in the dataset")
        
# Helper function to print top N similar movies
def print_similar_movies(movie_data, top_movie_id):
    print('Recommendations for the given movie: \n')
    for id in top_movie_id:   
        print(movie_data[movie_data.movieId == id].title.values[0])
```

Let's test our function by recommending the 10 most similar movies to One of my favorite movie "Heat". We can find that **most of the recommnended movies (9/10)** except 'Four Rooms' are Drama/Crime/Thriller/Action movies like "Heat', which indicates our recommendations engine works well. 

```python
recom_ids = top_cosine_similarity(vt, 'Heat', 10)
print_similar_movies(movies_data, recom_ids)
```

Then we test the case-insensitive input by typing 'heat' and we can still get the same result.

Then we define a function to obtain the information such as movie title, critics score, audience score and poster's url of the recommended movies.

```python
def get_movie_infor(movie_df, movie_id):
    indexes = []
    for id in movie_id:
        index = movie_df[movie_df['id']==id].index.astype(int)[0]
        indexes.append(index)
    get_content = movie_df.iloc[indexes]
    return get_content[['title','rtAllCriticsScore', 'rtAudienceScore', 'rtPictureURL']]
```

Lastly, we combine the previously defined functions to get the function we will use in our web app. 

```python
def recom_based_on_movie(low_dim_data, input_movie_name, top_n, movie_df):
    search_input = input_movie_name.upper()
    recom_ids = top_cosine_similarity(low_dim_data, search_input, top_n)
    print("Recommended Movies for: " + search_input)
    return get_movie_infor(movie_df, recom_ids)
```

## Recommend based on user's favorite genre

First load the Selectedmovies and Selectedmoviesgenres datasets we got from previous data gathering and preparation process.

```python
genres_df=pd.read_csv('C:/Users/xz2654/Downloads/Selectedmoviesgenres.csv', encoding='latin-1')
movie_df=pd.read_csv('C:/Users/xz2654/Downloads/selectedmovies_infor.csv')
```

Define a function recommend top 4 movies (4 highest audiences rated movies with low critics socre) based on user's favorite genre. The imput is one kind of movie genre while the output is top 5 movies' title, audience score, critics score and its poster's url. 

```python
def recom_based_on_genre(genres_df, movie_df, args='Comdedy', top_k=4, condition='rtAudienceScore'):
    get_id = genres_df[genres_df.genres.apply(lambda x: args in x.split('|'))].movieId
    get_content = movie_df[movie_df.id.apply(lambda x: x in list(get_id))].sort_values(condition, ascending=False).iloc[:top_k]
    return get_content[['title','rtAllCriticsScore', 'rtAudienceScore', 'rtPictureURL']]
```
