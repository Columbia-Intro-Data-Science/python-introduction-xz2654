## Problem 3:  Gradient Descent and the learning rate

By modifying the learning rate below, show how the convergence takes longer or doesn't converge at all.
Can you explain in words or math why this is?

````python
from numpy import *

# y = mx + b
# m is slope, b is y-intercept
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/N) * x * (y - ((m_current * x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
        b, m = step_gradient(b, m, array(points), learning_rate)
    return [b, m]

def run(num_iterations, learning_rate=0.0001):
    points = genfromtxt("/Users/xz2654/Downloads/data.csv", delimiter=",")
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    print("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
    print("Running...")
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1])
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')

run(100)
````

````python
# Use this to visually discuss convergence rate based on learning rate
for num in range(0,10):
    run(num)
    plt.show()
````

To get a big picture of how learning rate are related to the the convergence process. We set the iteration number to be 5 and choose different learning rate to get corresponding error after 5 iterations.

````python
initial_b = 0
initial_m = 0 
num_iterations = 5
errors=[]
learning_rates = np.logspace(-4,-2,100)

for l in learning_rates:
    learning_rate = l
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    errors.append(compute_error_for_line_given_points(b, m, points))

plt.figure(figsize=(10, 10))
plt.plot(learning_rates,errors)
plt.xlabel('Learning Rate')
plt.ylabel('Error ')
plt.show()
````

Based on the above plot, we can find the error starts to increases rapidyly if we set the learning rate greater than 0.006. The reason might be that we cannot not converge if we set the learning rate to be too large (in this case, larger than 0.006).

Then, we can see relationship between error and learning rate if we set the learning rate to be relatively small.

````python
initial_b = 0
initial_m = 0 
num_iterations = 5
errors=[]
learning_rates = np.logspace(-5,-4,100)

for l in learning_rates:
    learning_rate = l
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    errors.append(compute_error_for_line_given_points(b, m, points))

plt.figure(figsize=(10, 10))
plt.plot(learning_rates,errors)
plt.xlabel('Learning Rate')
plt.ylabel('Error ')
plt.show()
````

Based on the above plot, we can find the error after 5 iterations decrease as the learning rate increases. The reason might be that the converge processing would be very slow if we set the learning rate to be too small (in this case, smaller than 0.0001).

Conclusion: Based on the above plots, we can find that if we choose a learning rate that is too large (as in the first situation, we used 5), we might bypass the minimum point and cannot converege finally. On the other hand, if we choose a learning rate that is too small (as in the third situation, we used 0.000001), the coverge process then might be very slow and not efficient enough. Therefore, we must choose the learning rate very carefully.
